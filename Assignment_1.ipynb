````python
#3 Compare and contrast mutable and immutable objects in Python with examples.

-Ans:
      Mutable Objects:
      - Mutable objects can be modified after they are created. This means you can change their contents (e.g., update, add, or delete items) without changing the object's identity.
      - Examples: Lists, dictionaries, sets, user-defined objects (usually).
      -
        # Example of a mutable list
        my_list = [1, 2, 3]
        print(my_list)  # Output: [1, 2, 3]

        my_list.append(4)  # Modifying the list
        print(my_list)  # Output: [1, 2, 3, 4]

    Immutable Objects:
    - Immutable objects cannot be modified after they are created. Any operation that appears to modify the object will instead create a new object with the modified value.
    - Examples: Strings, tuples, integers, floats, and frozensets.
    -
     # Example of an immutable string
    my_string = "hello"
    print(my_string)  # Output: hello

    # Attempting to modify the string (this creates a new string)
    my_string = my_string + " world"
    print(my_string)  # Output: hello world

````

````python
#4 Discuss the different types of operators in Python and provide examples of how they are used.

-Ans:
      1. Arithmetic Operators:
        - Perform basic mathematical operations.
        - Example: `+`, `-`, `*`, `/`
        ```
        a = 5 + 2   # Output: 7
        ```

      2. Comparison Operators:
        - Compare values and return a boolean (`True` or `False`).
        - Example: `==`, `!=`, `>`, `<`
        ```
        5 == 5  # Output: True
        ```

      3. Logical Operators:
        - Combine conditional expressions.
        - Example: `and`, `or`, `not`
        ```
        True and False  # Output: False
        ```

      4. Assignment Operators:
        - Assign values to variables.
        - Example: `=`, `+=`, `-=`
        ```
        x = 10
        x += 5   # x becomes 15
        ```

      5. Bitwise Operators:
        - Perform operations on binary numbers.
        - Example: `&`, `|`, `^`, `~`, `<<`, `>>`
        ```
        5 & 3  # Output: 1
        ```

      6. Identity Operators:
        - Check if two variables refer to the same object.
        - Example: `is`, `is not`
        ```
        a = [1, 2]
        b = a
        a is b  # Output: True
        ```

      7. Membership Operators:
        - Test if a value is part of a sequence (like a list).
        - Example: `in`, `not in`
        ```
        3 in [1, 2, 3]  # Output: True
        ```
````

````python
#5 Explain the concept of type casting in Python with examples.

-Ans:
      Type casting is the process of converting one data type into another. Python supports two types of type casting:

      1. Implicit Type Casting:
        - Python automatically converts one data type to another during operations where necessary, without explicit instruction from the programmer.

        Example:
        ```
        a = 5      # int
        b = 2.5    # float
        c = a + b  # Implicitly converts 'a' to float
        print(c)   # Output: 7.5
        ```

      2. Explicit Type Casting:
        - The programmer manually converts one data type to another using built-in functions like `int()`, `float()`, `str()`, etc.

        Example:
        ```
        x = "10"          # string
        y = int(x)        # Converts to integer
        print(y + 5)      # Output: 15

        z = 3.14
        print(int(z))     # Converts float to integer, Output: 3
        ```

````

````python
#6 How do conditional statements work in python with examples.

-Ans:
     Conditional statements allow a program to execute different blocks of code based on certain conditions. The main conditional statements in Python are `if`, `elif`, and `else`.

      1. `if` Statement:
        Executes a block of code if the given condition is `True`.

        ```
        x = 10
        if x > 5:
            print("x is greater than 5")
        ```
        Output: `x is greater than 5`

      2. `else` Statement:
        Executes a block of code if the `if` condition is `False`.

        ```
        x = 2
        if x > 5:
            print("x is greater than 5")
        else:
            print("x is not greater than 5")
        ```
        Output: `x is not greater than 5`

      3. `elif` (Else if) Statement:
        Checks multiple conditions. If the previous conditions are `False`, it checks the next condition.

        ```
        x = 7
        if x > 10:
            print("x is greater than 10")
        elif x > 5:
            print("x is greater than 5 but less than or equal to 10")
        else:
            print("x is less than or equal to 5")
        ```
        Output: `x is greater than 5 but less than or equal to 10`

      4. Nested `if` Statements:
        You can place an `if` statement inside another `if` statement.

        ```
        x = 20
        if x > 10:
            print("x is greater than 10")
            if x > 15:
                print("x is also greater than 15")
        ```
        Output:
        ```
        x is greater than 10
        x is also greater than 15
        ```
````

````python
#7 Describe the different types of loops in Python and their use cases with examples.

-Ans:
      Python has two types of loops: `for` and `while`. Loops are used to execute a block of code multiple times until a condition is met.

      1. `for` Loop:
      - Used when you know the number of iterations or when you want to iterate over a sequence (like lists, tuples, dictionaries, strings).
      - Example:
        ```
        numbers = [1, 2, 3, 4, 5]
        for num in numbers:
            print(num)
        ```
        Output:
        ```
        1
        2
        3
        4
        5
        ```

      - Looping through a range of numbers.
        ```
        for i in range(1, 6):
            print(i)
        ```
        Output:
        ```
        1
        2
        3
        4
        5
        ```

      2. `while` Loop:
      - Used when the number of iterations is not known beforehand, and you want to keep looping until a condition becomes `False`.
      - Example:
        ```
        i = 1
        while i <= 5:
            print(i)
            i += 1
        ```
        Output:
        ```
        1
        2
        3
        4
        5
        ```
````
